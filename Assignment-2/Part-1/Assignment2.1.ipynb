# Step 4: Aggregation using GROUP BY

Aggregations summarize large amounts of data into meaningful business logics.

### # Total Sales by Category

```sql
select Category,round(sum(Sales),2) as Total_Sales
from sample_superstore
group by Category
order by  Total_Sales desc;
```

I am running this query to group our data by product category and calculate the total sales for each, sorted from highest to lowest so I can instantly see which category brings in the most money.rounding it to 2 decimal points as its giving lot of decimal values;

### Total Quantity Sold by Category

```sql
select Category,SUM(Quantity) as Total_Quantity
from sample_superstore
group by Category
order by Total_Quantity desc;
```

I am running this query to group our data by product category and count the total number of items sold for each, sorted from highest to lowest so I can instantly see which category is our most popular in terms of volume.

### Average Sales by Category

```sql
select Category,round(avg(Sales),2) as Average_Sales
from sample_superstore
group by Category;
```

Average sales help compare the typical order value across categories.

# Step 5: Sorting and Limiting Results

ORDER BY and LIMIT help identify top-performing products and categories.

### Top 10 Products

```sql
select `Product Name`,round(sum(Sales),2) as Total_Sales
from sample_superstore
group by `Product Name`
order by Total_Sales desc
limit 10;
```

These are the highest revenue-generating products.

### Top Categories

```sql
select Category,ROUND(SUM(Sales),2) as Total_Sales
from sample_superstore
group by Category
order by Total_Sales desc
limit 3;
```

The top categories contribute the majority of overall sales revenue.

# Step 6: Business Use Cases

The following queries solve real-world business problems.

### Monthly Sales Trends

```sql
select year(`order date`) as year,month(`order date`) as month,round(sum(sales), 2) as monthly_sales
from sample_superstore
group by year(`order date`),month(`order date`)
order by year, month;
```

showing every year monthwise sales analysis

### # Top Customers

```sql
select `customer name`,round(sum(sales), 2) as total_sales
from sample_superstore
group by `customer name`
order by total_sales desc
limit 10;
```

### Duplicate Records

```sql
select`order id`,`product id`,count(*) as duplicate_count
from sample_superstore
group by `order id`,`product id`
having count(*) > 1;
```

# Step 7: Data Validation

### Total Row Count

```sql
select count(*) as total_rows from sample_superstore;
```

### Check Missing Customer IDs

```sql
select count(*) - count(`customer id`) as missing_customer from sample_superstore;
```

count(*) gives all rows count and count(`customer id`) gives count of rows where customer id is not null

### Check Missing Product IDs

```sql
select count(*) - count(`product id`) as missing_products
from sample_superstore;
```

### Check Missing Sales Values

```sql
select count(*) - count(sales) as missing_sales
from sample_superstore;
```
