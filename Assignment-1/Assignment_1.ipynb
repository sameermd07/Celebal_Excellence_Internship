## Task-4(Performing basic operations (filter rows, select columns))

```python
csv_1 = csv_1[
    (csv_1["rating"] >= 0) &
    (csv_1["rating"] <= 5)
]
```

```python
csv_1 = csv_1[
    (csv_1["discount"] >= 0) &
    (csv_1["discount"] <= 100)
]
```

```python
csv_1 = csv_1[csv_1["ratings_count"] >= 0]
```

```python
csv_1 = csv_1[csv_1["title"].str.strip() != ""]
```
